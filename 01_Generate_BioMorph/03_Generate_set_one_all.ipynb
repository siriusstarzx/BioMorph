```python
import pandas as pd
```

```python
results = pd.read_csv("Linear_combinations_Cell_Health.csv")
results = results[results["r2"]>0.25].reset_index(drop=True)
results
```

```python
from ast import literal_eval
results.selected_features = results.selected_features.apply(literal_eval)
```

```python
cp_scaled = pd.read_csv("cp_scaled.csv")
cp_scaled
```

```python
endpoint_metadata = pd.read_csv("endpoint_metadata_cell_process.csv")
endpoint_metadata
```

```python
from sklearn.preprocessing import LabelBinarizer
```

```python
from sklearn.ensemble import RandomForestClassifier
from boruta import BorutaPy
```

```python
from sklearn.model_selection import train_test_split


    
```

```python
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score
from sklearn.metrics import classification_report
from sklearn.metrics import balanced_accuracy_score
from sklearn.metrics import matthews_corrcoef,confusion_matrix, roc_auc_score, roc_curve
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_predict
from numpy import argmax

from sklearn.ensemble import RandomForestClassifier
from boruta import BorutaPy
```

```python
results_metrics = pd.DataFrame()
list_of_lists = []

for endpoint in results.endpoint.to_list()[:]:
    print(endpoint)
    
    features = results[results["endpoint"]==endpoint].reset_index()["selected_features"][0]
    
    print("Number of overall selected features for this cell health endpoint:")
    print(len(features))
    if(len(features)<1):
        continue
        
    #CP dataset with only selected features
    cp_scaled_selected = cp_scaled[features]
    cp_scaled_selected["Metadata_profile_id"]= cp_scaled["Metadata_profile_id"]
    
    cp_scaled_selected_2 = pd.merge(cp_scaled_selected, endpoint_metadata)

    neg_ctrl ="Negative Ctrl"
    cp_scaled_selected_neg_control = cp_scaled_selected_2[cp_scaled_selected_2["cell_process"]==neg_ctrl].reset_index(drop=True)
    cp_scaled_selected_neg_control["cell_process"] = str(neg_ctrl)#+"_"+cp_scaled_selected_neg_control["Metadata_profile_id"]
    cp_scaled_selected_neg_control_columns= list(cp_scaled_selected_neg_control["cell_process"].unique())
    
    print("Number of Negative controls for this cell health endpoint:")
    
    len_cp_scaled_selected_neg_control= len(cp_scaled_selected_neg_control)
    print(len_cp_scaled_selected_neg_control)
    
    if(len_cp_scaled_selected_neg_control<1):
        continue
        
    #Cell Processes which are not negative control
    
    cell_processes = list(cp_scaled_selected_2.cell_process.unique())  
    if "Negative Ctrl" in cell_processes: cell_processes.remove("Negative Ctrl")
    if np.nan in cell_processes: cell_processes.remove(np.nan)
    
    print("Number of Cell Processes (besides negative control) for this cell health endpoint:")
    print(len(cell_processes))
    
    for cell_process in cell_processes:
        print(cell_process)
        
        cp_scaled_selected_cell_process = cp_scaled_selected_2[cp_scaled_selected_2["cell_process"]==cell_process].reset_index(drop=True)
        cp_scaled_selected_cell_process["cell_process"] = str(cell_process)#+"_"+cp_scaled_selected_cell_process["Metadata_profile_id"]
        cp_scaled_selected_cell_process_columns= list(cp_scaled_selected_cell_process["cell_process"].unique())
        
        len_cp_scaled_selected_cell_process= len(cp_scaled_selected_cell_process)
        print("Number of perturbations for this cell process:")
        print(len_cp_scaled_selected_cell_process)
        
        if(len_cp_scaled_selected_cell_process<1):
            continue
            
        #neg Control and cell process
        cp_scaled_selected_cell_process_neg_control = pd.concat([cp_scaled_selected_neg_control,cp_scaled_selected_cell_process]).reset_index(drop=True)

        X = cp_scaled_selected_cell_process_neg_control[features].values
        y = cp_scaled_selected_cell_process_neg_control["cell_process"].values
        ohe = LabelBinarizer()
        y = ohe.fit_transform(y[:, np.newaxis])
        y = y.ravel()

        clf = RandomForestClassifier(n_jobs=1, max_depth=5, random_state=42)
        feat_selector = BorutaPy(clf, verbose=1, random_state=42)
        feat_selector.fit(X, y)

        #Find features imp for the cell health and cell process combination
        features_df= pd.DataFrame(features, columns=["features"])
        features_df["rank"]=feat_selector.ranking_
        features_df["Confirmed"]=feat_selector.support_
        selected_features = features_df[features_df["Confirmed"]==True].features.to_list()
        
        
        print("Number of selected features for this cell process and cell health endpoint:")
        print(len(selected_features))
        
        if(len(selected_features)<1):
            continue
        
        
        #Run RF models to check if selection is meaningful
        X = cp_scaled_selected_cell_process_neg_control[selected_features].values
        y = cp_scaled_selected_cell_process_neg_control["cell_process"].values
        ohe = LabelBinarizer()
        y = ohe.fit_transform(y[:, np.newaxis])
        y = y.ravel()
        
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42) 
        
        #Find best threshold
        inner_cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
        try:
            cross_val_prob = cross_val_predict(RandomForestClassifier(n_jobs=-1, max_depth=5, random_state=42), X_train, y_train, cv=inner_cv, method='predict_proba')[:, 1]

            fpr, tpr, thresholds = roc_curve(y_train, cross_val_prob)
                    # get the best threshold
            J = tpr - fpr
            ix = argmax(J)
            best_thresh = thresholds[ix]
        except: 
            best_thresh =0.50
           
        print('Best Threshold=%f' % (best_thresh))
        
        clf = RandomForestClassifier(n_jobs=-1, max_depth=5, random_state=42)
        clf.fit(X_train, y_train)

        #predicitons
        y_prob=  clf.predict_proba(X_test)[:,1]
        y_pred  = [ 1 if x>best_thresh  else 0 for x in y_prob ] 
        y_true = y_test
        
        
        #Metrics
        conf_matrix = confusion_matrix(y_true, y_pred)
        print(conf_matrix)

        print(classification_report(y_true, y_pred))

        ba= balanced_accuracy_score(y_true, y_pred)
        print('balanced_accuracy_score ',ba)

        try: mcc=matthews_corrcoef(y_true, y_pred)
        except: mccc=-2
            
        print('matthews_corrcoef ',mcc)
        # Sensitivity, hit rate, recall, or true positive rate


        try: Specificity = conf_matrix[0,0]/(conf_matrix[0,0]+conf_matrix[0,1])
        except: Specificity=-2
        # Specificity or true negative rate
        try: Sensitivity = conf_matrix[1,1]/(conf_matrix[1,0]+conf_matrix[1,1]) 
        except: Sensitivity=-2
            
        print( 'Sensitivity', Sensitivity)
        print( 'Specificity', Specificity)
        AUC = roc_auc_score(y_true, y_prob)
        print('AUC-ROC ',AUC)    
        
        row= [endpoint, cell_process, selected_features, ba, AUC, mcc, len_cp_scaled_selected_neg_control, len_cp_scaled_selected_cell_process]
        list_of_lists.append(row)
        
        results_metrics = pd.DataFrame(list_of_lists, columns=["endpoint", "cell_process", "selected_features", 
                                               "ba", "AUC", "mcc", "len_cp_scaled_selected_neg_control", 
                                               "len_cp_scaled_selected_cell_process"])

        #results_metrics.to_csv("Cell_Painting_ontology_v3_BAK.csv", index=False)
```
