```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.model_selection import StratifiedKFold
```

```python
from ast import literal_eval
import numpy as np

df= pd.read_csv("Cell_Painting_ontology_v3_selected_ontology.csv")
df["CP_Ontology"]=df["feature_type"]+str("_")+df["measurement"]+str("_")+df["endpoint"]+str("_")+df["cell_process"]
df= df[["CP_Ontology", "selected_features"]]
df.selected_features = df.selected_features.apply(literal_eval)
df
```

```python
from scipy.stats import chisquare
```

```python
for CP_Ontology in df.CP_Ontology.to_list():
    #print(CP_Ontology)
    
    features_from_ontology = list(set(df[df["CP_Ontology"]==CP_Ontology].reset_index()["selected_features"][0]))
    #print(features_from_ontology)
    
    CP[CP_Ontology] = chisquare(CP[features_from_ontology].T.values)[1]
```

```python
CP_new_dataset= CP[df.CP_Ontology.to_list()]
CP_new_dataset["InChICode_standardised"]= CP["InChICode_standardised"]
CP_new_dataset
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
print(scaler.fit(CP_new_dataset.iloc[:, :-1]))
```

```python
CP_new_dataset_scaled= pd.DataFrame(scaler.transform(CP_new_dataset.iloc[:, :-1]), columns=CP_new_dataset.iloc[:, :-1].columns.to_list())
CP_new_dataset_scaled = CP_new_dataset_scaled.replace(np.inf, np.nan)
CP_new_dataset_scaled = CP_new_dataset_scaled.replace(-np.inf, np.nan)
CP_new_dataset_scaled.dropna(axis=1, inplace=True)
CP_new_dataset_scaled["InChICode_standardised"]= CP["InChICode_standardised"]
CP_new_dataset_scaled
```
