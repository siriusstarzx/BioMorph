```python
import pandas as pd
```

```python
df1 = pd.read_csv("metrics_df_CP_multiple_runs.csv")
#df1
```

```python
df2 = pd.read_csv("metrics_df_Ontology_pvalue_multiple_runs.csv")
#df2
```

```python
df3 = pd.read_csv("metrics_df_Ontology_statistic_multiple_runs.csv")
#df3
```

```python
metrics_list = pd.concat([df1, df2, df3]).reset_index(drop=True)
#metrics_list
```

```python
metrics_list.labels_name.unique()
```

```python
metrics_list
```

```python
metrics_list.groupby(["labels_name", "Features"]).mean()
```

```python
metrics_list.groupby(["Features"]).mean()
```

```python
metrics_list = metrics_list.replace({'Ontology_pvalue': "Cell Health terms"})
metrics_list = metrics_list.replace({'CP_features': "Cell Painting features"})
metrics_list
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from statannotations.Annotator import Annotator

pairs = [('Cell Health terms', 'Cell Painting features')]

my_order=['Cell Health terms', 'Cell Painting features']
         

fig, ax = plt.subplots(1, 3,figsize=(20, 4), dpi=300) # sharex='all', sharey='all', 

for i, endpoint in enumerate(['apoptosis up', 'cytotoxicity BLA', 'cytotoxicity SRB']):
    print(i)
    print(endpoint)

    sns.violinplot(ax= ax[i], x="Features", y='AUC', data=metrics_list[metrics_list["labels_name"]==endpoint], order=my_order, scale='width', inner='quartile', color="white")
    sns.swarmplot(ax= ax[i],x='Features', y='AUC', data=metrics_list[metrics_list["labels_name"]==endpoint], order=my_order,color="grey", alpha=0.8)
    annotator = Annotator(ax[i], pairs, data=metrics_list[metrics_list["labels_name"]==endpoint], x="Features", 
                          y='AUC', order=my_order)
    annotator.configure(test='t-test_paired', text_format='star', loc='outside')
    annotator.apply_and_annotate()
    
plt.show()


```
