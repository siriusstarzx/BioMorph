```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.model_selection import StratifiedKFold
```

```python
CP = pd.read_csv('data/CP_data_Mitotox_Mechanisms_658_827features_scaled.csv') 
CP
```

```python
CP_features_list = CP.iloc[:, :-1].columns.to_list()
len(CP_features_list)
```

```python
labels = pd.read_csv('data/Endpoints_9_Mitotox_biological_processes_658.csv')
labels_list= labels.iloc[:, 1:].columns.to_list()
labels_list
```

```python
df = pd.merge(CP, labels, on='InChICode_standardised')
df
```

```python
from sklearn import metrics
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import balanced_accuracy_score
import matplotlib.pyplot as plt
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint
from numpy import argmax
from sklearn.experimental import enable_halving_search_cv
from sklearn.model_selection import HalvingRandomSearchCV
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_predict
```

```python
def to_labels(pos_probs, threshold):
    return (pos_probs >= threshold).astype('int')
```

```python
skf = StratifiedKFold(n_splits=4, shuffle=True, random_state=42)

list_of_list = []
compound_pred_list=[]

for labels_name in labels_list[:]:
    print(labels_name)
    
    df_assay = df.dropna(subset=[labels_name]).reset_index(drop=True)
    print(df_assay[labels_name].value_counts())
    
    count = 0
    for train_index_outer, test_index in skf.split(df_assay, df_assay[labels_name]):
        
        assay_results = pd.DataFrame()
        val_results = pd.DataFrame()

        count += 1
        save_name = f"{labels_name}_{count}"
        print(f"\n ***{save_name}***\n")
        
        train, test = df_assay.iloc[train_index_outer], df_assay.iloc[test_index]
        
        print("len(Train)", len(train))
        print("len(Test)", len(test))
    
        compounds_train = train["InChICode_standardised"].to_numpy() 
        X_train= train[CP_features_list].to_numpy()
        compounds_test = test["InChICode_standardised"].to_numpy() 
        X_test= test[CP_features_list].to_numpy()
        
        y_train= train[labels_name].to_numpy()
        y_test= test[labels_name].to_numpy()
        
        print("shape(X_Train)", (X_train.shape))
        print("shape(y_Train)", (y_train.shape))
        
        #break
        
        rf = RandomForestClassifier(n_jobs=-1, random_state=42)
    

        # Instantiate the RandomHalving search model
        param_dist_grid = { 
        'max_depth': randint(5, 20),
        'max_features': randint(20, 150),
        'min_samples_leaf': randint(5, 15),#
        'min_samples_split': randint(5, 15),#
        'n_estimators':[200, 400, 600],#
        'bootstrap': [True, False],#
        'oob_score': [False],
        'random_state': [42],
        'criterion': ['gini', 'entropy'],#
        'n_jobs': [-1],
        'class_weight' : [None, 'balanced']#
            }

        inner_cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)    
        rsh = HalvingRandomSearchCV(estimator=rf, param_distributions=param_dist_grid,
                                    factor=2, random_state=42, n_jobs=-1, verbose=1, cv = inner_cv)



        rsh.fit(X_train, y_train)
        classifier = rsh.best_estimator_
        classifier.fit(X_train, y_train)
        
        test_proba = classifier.predict_proba(X_test)[:,1]
        test_pred = to_labels(test_proba, 0.50)
        ba_old=balanced_accuracy_score(y_test, test_pred) 
        print("Old BA", ba_old.round(2))
        
        fpr, tpr, thresholds = metrics.roc_curve(y_test, test_proba)
        AUC = metrics.auc(fpr, tpr)
        print("AUC", AUC.round(2))
        
        cross_val_prob = cross_val_predict(classifier, X_train, y_train, cv=inner_cv, method='predict_proba')[:, 1]
        fpr, tpr, thresholds = metrics.roc_curve(y_train, cross_val_prob)
                    # get the best threshold
        J = tpr - fpr
        ix = argmax(J)
        best_thresh = thresholds[ix]
        print('Best Threshold=%f' % (best_thresh))
        
        test_preds_labels = to_labels(test_proba, best_thresh)
        ba_new=balanced_accuracy_score(y_test, test_preds_labels)
        print("New BA", ba_new.round(2))
        print(" ")
        
        importances = classifier.feature_importances_
        
        filtered_CP_imp_features_list = np.array(CP_features_list)[np.array(importances>(importances.mean() + 2*importances.std()))]
        print("Filtered_CP_len:", len(set(filtered_CP_imp_features_list)))
        
        row= [labels_name, "CP_features", AUC, ba_new, list(filtered_CP_imp_features_list)]
        list_of_list.append(row)
        
        metrics_df= pd.DataFrame(list_of_list, columns=["labels_name", "Features", "AUC",  "ba", "Filtered_CP_imp_features"])
        #metrics_df.to_csv("metrics_df_CP_RF_forest_importances_allendpoints_singlerun.csv", index=False)
        
        break

    predictions= [X_compound_test, y_test, test_preds_labels]
    predictions= pd.DataFrame(predictions, index=["InChICode_standardised", "true", "pred"]).T
    predictions["labels_name"]=labels_name
    predictions = predictions[predictions["true"]==1]
    predictions = predictions[predictions["pred"]==1]
    
    print("True Positives: ", len(predictions))

    for i in predictions.index:

        #print(X_compound_test[i])
        explainer = shap.TreeExplainer(classifier)
        shap_values = explainer.shap_values(X_test[i])

        shap_df = pd.DataFrame([CP_features_list, shap_values[1]], index=["feature", "SHAP"]).T.sort_values(by="SHAP", ascending=False)
        shap_df = shap_df[shap_df["SHAP"]>0].reset_index(drop=True)

        filtered_SHAP_imp_features_list = shap_df[shap_df["SHAP"]>(shap_df.SHAP.mean()+ 2*shap_df.SHAP.std())].feature.to_list()
        #print("Filtered_CP_SHAP_imp_len:", len(set(filtered_SHAP_imp_features_list)))

        row_r=[labels_name, X_compound_test[i], filtered_SHAP_imp_features_list]
        compound_pred_list.append(row_r)

        Preds_SHAP_df= pd.DataFrame(compound_pred_list, columns=["labels_name", "InChICode_standardised", "Filtered_CP_SHAP_features"])
        #Preds_SHAP_df.to_csv("metrics_df_CP_RF_preds_SHAP_allendpoints_singlerun.csv", index=False)

#Check no feature had zero importance
```
