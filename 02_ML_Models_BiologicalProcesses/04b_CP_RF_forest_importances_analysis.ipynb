```python
import math

# Calculate the number of rows and columns for subplot
n = len(df.labels_name.unique())
cols_subplot = 1
rows_subplot = min(n, 8) # The number of rows will be the minimum of n and 8

fig, axs = plt.subplots(rows_subplot, cols_subplot, figsize=(10, 14), dpi=300)

# If there's only one row, axs is not a list. Convert it to list.
if rows_subplot == 1:
    axs = [axs]

plt.rcParams.update({'font.size': 16})

idx = 0 # to keep track of the subplot

for label in df.labels_name.to_list():
    if(df[df["labels_name"]==label]["AUC"].values <=0.50):
        continue
    
    print(label)
    
    ontology_enrichment = Check_Ontology.return_ontology(set(df[df["labels_name"]==label].reset_index()["Filtered_CP_imp_features"][0]))

    # Endpoint wise analysis
    endpoint_wise = ontology_enrichment.groupby(["endpoint"]).mean()#[ontology_enrichment.groupby(["endpoint"]).mean()["cell_process_perc_overlap"]> 5]
    endpoint_wise = endpoint_wise.sort_values(by =["cell_process_perc_overlap"], ascending = False).reset_index()
    endpoint_wise = endpoint_wise.iloc[:5]

    barplot = sns.barplot(data=endpoint_wise, x="cell_process_perc_overlap", y="endpoint", lw=1, ax=axs[idx])

    # Loop over the bars, and assign a color and label to each
    for i, bar in enumerate(barplot.patches):
        barplot.plot([], [], color=bar.get_facecolor(), label=endpoint_wise.iloc[i]['endpoint'])

    barplot.set_title(' ')
    barplot.set_xlim(0, 72)
    if idx == 7:
        barplot.set_xlabel('Cell Health phenotypes Percentage Overlap (%)')
    else:
        barplot.set_xlabel('')
        barplot.set_xticklabels([])
    barplot.set_ylabel(' ')
            
    barplot.bar_label(barplot.containers[0], fmt='%.1f')
    barplot.tick_params(bottom=False) 

    idx += 1

    # Stop creating plots after reaching 8 rows
    if idx == 8:
        break

plt.tight_layout()
plt.show()
```
