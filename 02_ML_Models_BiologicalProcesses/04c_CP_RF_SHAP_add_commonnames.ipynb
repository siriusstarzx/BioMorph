```python
import pandas as pd
```

```python
df = pd.read_csv("metrics_df_CP_RF_preds_SHAP_allendpoints_singlerun.csv")
df
```

```python
from ast import literal_eval
import numpy as np

df.Filtered_CP_SHAP_features = df.Filtered_CP_SHAP_features.apply(literal_eval)
df
```

```python
from rdkit.Chem import inchi
from rdkit import Chem

def get_standardize_smiles(value):
    try: return Chem.MolToSmiles(Chem.inchi.MolFromInchi(value))
    except: return "Cannot_do"
    
from pandarallel import pandarallel
pandarallel.initialize()

df['smiles_r']=df['InChICode_standardised'].parallel_apply(get_standardize_smiles)  
df
```

```python
import pubchempy

def GetCommonName(smiles):

    compounds = pubchempy.get_compounds(smiles, namespace='smiles')
    match = compounds[0]
    return (match.synonyms[0])

def GetIUPAC(smiles):

    compounds = pubchempy.get_compounds(smiles, namespace='smiles')
    match = compounds[0]
    return (match.iupac_name)
```

```python
df["iupac_name"] = df["smiles_r"].apply(GetIUPAC)
```

```python
df
```

```python
df["Common_name"] = df["smiles_r"].apply(GetCommonName)
```
